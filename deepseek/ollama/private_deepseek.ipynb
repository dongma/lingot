## 调用私有部署的deepseek模型

```python
import os
from openai import OpenAI

# 初始化 DeepSeek 客户端
client = OpenAI(
    api_key="ollama", # 本地部署随便写，非空即可
    base_url="http://10.10.0.47:11434/v1",  # DeepSeek API 的基地址
)
# 让deepseek-r1:8b生成环保咖啡杯文案
stream = client.chat.completions.create(
    model="deepseek-r1:8b",
    messages=[{"role": "user", "content": "生成一个环保咖啡杯的小红书文案"}],
    stream=True
)

for chunk in stream:
    print(chunk.choices[0].delta.content or "", end="", flush=True)
```
